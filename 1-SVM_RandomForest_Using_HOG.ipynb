## Classification using Support vector Machine and Random by incorporating Histogram of oriented gradient

```python
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from sklearn.svm import LinearSVC
import os
import numpy as np
import cv2
import pandas as pd
import matplotlib.pyplot as plt
```

### Inria Dataset path

```python
person_train_images = './INRIA_Dataset_Samples/Train/pos'
non_person_train_images = './INRIA_Dataset_Samples/Train/neg'
person_test_images = './INRIA_Dataset_Samples/Test/pos'
non_person_test_images = './INRIA_Dataset_Samples/Test/neg'
```

#### Training data and labels array

```python
X_train = []
y_train = []
```

### Do processing for positive instances

```python
# Get all postive train person images
for image in os.listdir(person_train_images):
    if image.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')):
        print("processing =  {}".format(image))
        img = imread(person_train_images+"/"+image)
        resized_img = resize(img, (128 * 4, 64 * 4))
        histogram = hog(resized_img, orientations=8, pixels_per_cell=(16, 16),cells_per_block=(1, 1), visualize=False)
        X_train.append(histogram)
        y_train.append(1)
```

#### Do Processing for Negative images

```python
# Get all negative train images
for image in os.listdir(non_person_train_images):
    if image.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')):
        print("processing =  {}".format(image))
        img = imread(non_person_train_images+"/"+image)
        resized_img = resize(img, (128 * 4, 64 * 4))
        histogram = hog(resized_img, orientations=8, pixels_per_cell=(16, 16), cells_per_block=(1, 1), visualize=False)
        X_train.append(histogram)
        y_train.append(0)
```

#### Converting extracting list of samples and labels into numpy array

```python
X_train = np.float32(X_train)
y_train = np.array(y_train)

print("Training dataset dimensions  = {}".format(X_train.shape))
print("Test dataset dimensions   = {}".format(y_train.shape))
```

## Start Training Using Support vector machine

```python
from sklearn.preprocessing import StandardScaler
from sklearn import svm

classifier = svm.SVC(gamma=0.001)

classifier.fit(X_train,y_train)
```

```python
X_test=[]
Y_test=[]
# Get all positve test images
for image in os.listdir(person_test_images):
    if image.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')):
        print("processing =  {}".format(image))
        img = imread(person_test_images+"/"+image)
        resized_img = resize(img, (128 * 4, 64 * 4))
        histogram = hog(resized_img, orientations=8, pixels_per_cell=(16, 16),cells_per_block=(1, 1), visualize=False)
        X_test.append(histogram)
        Y_test.append(1)
        

# Get all negative test images
for image in os.listdir(non_person_test_images):
    if image.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')):
        print("processing =  {}".format(image))
        img = imread(non_person_test_images+"/"+image)
        resized_img = resize(img, (128 * 4, 64 * 4))
        histogram = hog(resized_img, orientations=8, pixels_per_cell=(16, 16),cells_per_block=(1, 1), visualize=False)
        X_test.append(histogram)
        Y_test.append(0)
```

```python
test_accuracy = classifier.score(X_test, Y_test)
print(test_accuracy)
```

```python
from sklearn.preprocessing import StandardScaler
test_accuracy = []
scaler = StandardScaler()
```

```python
y_pred = classifier.predict(X_test)
```

```python
classes=["person","non_person"]
```

```python
from sklearn.metrics import f1_score

f1_score(Y_test, y_pred, average='micro')
```

```python
from sklearn import metrics

quantitative_Result = pd.DataFrame(metrics.confusion_matrix(Y_test, y_pred),columns = classes )
quantitative_Result.index = [classes]
quantitative_Result
```
